```python
import pandas as pd
import numpy as np 
```

```python
data = pd.read_csv('world_data.csv')
data.info()
```

```python
data.head()
```

```python
data.rename(columns={'Date_reported' : 'Date'}, inplace=True)
data.drop(['Country_code', 'WHO_region'], axis=1, inplace=True)
data.head()
```

```python
data.to_csv('World_data.csv', index=False)
```

```python
fr_data = data.loc[data.Country == 'France'].reset_index(drop=True)
fr_data.to_csv('france.csv', index=False)
fr_data.head()
```

```python
wn_cases = pd.DataFrame(data, columns = ['Date', 'New_cases'])
wn_cases.to_csv('world_new_cases.csv', index=False)
wn_cases.head()
```

```python
wc_cases = pd.DataFrame(data, columns = ['Date', 'Cumulative_cases'])
wc_cases.to_csv('world_cum_cases.csv', index=False)

fn_cases = pd.DataFrame(fr_data, columns = ['Date', 'New_cases'])
fn_cases.to_csv('fr_new_cases.csv', index=False)

fc_cases = pd.DataFrame(fr_data, columns = ['Date', 'Cumulative_cases'])
fc_cases.to_csv('fr_cum_cases.csv', index=False)
```

```python
wn2_cases = pd.DataFrame(data, columns = ['Date', 'New_deaths'])
wc_cases.to_csv('world_new_deaths.csv', index=False)

wc2_cases = pd.DataFrame(data, columns = ['Date', 'Cumulative_deaths'])
wn_cases.to_csv('world_cum_deaths.csv', index=False)

fn2_cases = pd.DataFrame(fr_data, columns = ['Date', 'New_deaths'])
wn_cases.to_csv('fr_new_deaths.csv', index=False)

fn2_cases = pd.DataFrame(fr_data, columns = ['Date', 'Cumulative_deaths'])
wn_cases.to_csv('fr_cum_deaths.csv', index=False)
```
